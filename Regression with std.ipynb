```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Hydroxylysine OI data.csv")
df
```

```python
df.columns
```

```python
df_clean = df[[
    'Cell1',
    'Age',
    'Sex',
    'Type',
    'n',
    'Method',
    '% hydroxylysine',
    'STD1']]
df_clean
```

```python
df_clean.isna().sum()
```

```python
df_clean.Type = df_clean.Type.astype(int)
df_clean.Method = df_clean.Method.astype(int)
df_Type_dummies = pd.get_dummies(df_clean.Type, prefix='Type')
df_Method_dummies = pd.get_dummies(df_clean.Method, prefix='Method')
df_clean.join(df_Type_dummies).join(df_Method_dummies)
```

```python
#df_train = df_clean[df_clean.STD1 != 0]
#df_train
```

```python
df_train = df_clean.dropna(subset=['STD1'])
df_train
```

```python
df_clean
```

```python
X = df_train[[
  'Cell1',
  'Age',
  'Sex',
  'Type',
  'n',
  'Method',
  '% hydroxylysine'
]]
X
```

```python
Y = df_train[['STD1']]
Y
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X = X, y = Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
df_clean
df_clean['STD1'] = df_clean['STD1'].fillna(0)
df_clean['n'] = df_clean['n'].replace(0,1)
df_clean
```

```python
df_test = df_clean[(df_clean.STD1 == 0) & (df_clean.n == 1)]
df_test
```

```python
new_X = df_test[[
  'Cell1',
  'Age',
  'Sex',
  'Type',
  'n',
  'Method',
  '% hydroxylysine'
]]
new_X
```

```python
from google.colab import files
reg.predict(new_X)
w=reg.predict(new_X)
arr = np.array(w)
print(type(arr))
np.savetxt("my_array.csv", arr, delimiter=",")
files.download('my_array.csv')
```
